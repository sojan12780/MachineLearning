Here we dont have null values but we have zeros. SOo, how do we find this values and how do we remove this?


```python
cols = ['Glucose','BloodPressure','SkinThickness', 'BMI','Insulin']
for col in cols:
  df[col]= df[col].replace(0,np.NaN)
  df[col] = df[col].fillna(df[col].mean())
```

```python
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .2)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors =13, metric = 'euclidean' )
```

```python
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
