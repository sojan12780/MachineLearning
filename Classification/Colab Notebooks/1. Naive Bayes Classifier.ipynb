```python
x = df.iloc[:,0:-1]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.4, random_state=1)
```

```python
x_train.shape
```

```python
x_train
```

## There are three kind of Naive bayes classifier:

    1. GaussianNB: Feature types should be Continous(random values, which may not be whole number)

    2. MultinomialNB: Feature types should be Discreate(Finite, Numeric, countable and non negative integer=>2.10,15,20)

    3. BernoulliNB:Feature types should be Binary(0 or 1, True or False, Yes or No)

```python
from sklearn.naive_bayes import GaussianNB
gnb= GaussianNB()
gnb.fit(x_train,y_train)
```

## There are four types function

    1. fit => fit(x_train, y_train) => make our model to learn something

    2. predict => model learn from traing part and use that in prediction part

    3. transform => We usually use it in the terms of transforming our data

    4. . fit_transform => We usually use it in the terms of transforming our data but here it lern from train and use it for transform the data in testing

```python
pred = gnb.predict(x_test)
```

## What can be a method to chek how well are working our dataset?

    We have few matrics to perform and check the accurecuy. How we can apply it? => we can call a matrics from sklearn.


```python
from sklearn import metrics 
metrics.accuracy_score(y_test,pred)
```

## Let's check about the confusion matrix

confusion matrix.png

```python
metrics.confusion_matrix(y_test,pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()

mnb.fit(x_train,y_train)
```

```python
pred = mnb.predict(x_test)
```

```python
metrics.accuracy_score(y_test,pred)
```

```python
from sklearn.naive_bayes import BernoulliNB
mnb = BernoulliNB()

mnb.fit(x_train,y_train)
```

```python
pred = mnb.predict(x_test)
```

```python
metrics.accuracy_score(y_test,pred)
```
