## Now Spliting dataset to X and Y

```python
x =df.iloc[:,0:-1]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
y.value_counts().plot.pie()
```

## Now mapping the necessary dataset

```python
cols = ['job',	'marital',	'education',	'default','month',	'housing',	'loan'	]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cols:
  x[col]=le.fit_transform(x[col])
```

```python
x
```

```python
#le.transform(x_train['education']) => for mapping one by one
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y = le.fit_transform(y)
```

```python
y
```

## And now splinting the dataset to Test and train using sklearn

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

dc = DecisionTreeClassifier()
dc.fit(x_train,y_train)
```

## Checking whather the model is overfited

```python
pred = dc.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_train,pred)
```

## Checking the accuracy

```python
pred = dc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,pred)
```

## Calling random forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,pred)
```
